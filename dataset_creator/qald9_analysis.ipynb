```python
import json
import pandas as pd
```

```python
with open("q9.json","r") as f:
    js = json.loads(f.read())
```

```python
js = js["questions"]
```

```python
js
```

```python
js[0]["id"][0]
```

```python
question = []
ids=[]
lang = []
for j in js:
    id=j["id"]
    q=j["question"]
    for l in q:
        if "string" in l:
            question.append(l["string"])
            ids.append(id)
            lang.append(l["language"])
        elif "string" not in l:
            continue
        
```

```python
output = pd.DataFrame({"questionId":ids,'questions': question, 'lang': lang})
output.to_csv('qald-9-train.csv', index=False)
```
